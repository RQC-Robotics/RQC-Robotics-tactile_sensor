```python
import os
from os.path import join as jn
```

```python
ds_path = 'data/videos_from_isaac/'
leave_number = 20
```

```python
print(os.listdir(ds_path))
```

```python
ids = os.listdir(ds_path)
for id in ids[:3]:
    item_path = jn(ds_path, id)
    for video in os.listdir(item_path):
        number = int(video)
        if number >= leave_number:
            print("removing ", jn(item_path, video, 'force.npz'))
            os.remove(jn(item_path, video, 'force.npz'))
            os.removedirs(jn(item_path, video))
        else:
            print("Not removing ", jn(item_path, video, 'force.npz'))
```
